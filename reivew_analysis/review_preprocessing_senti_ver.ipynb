```python
# 여러 단어 기준으로 문자열 분리하는 함수 정의
def split_by_multiple_words(text, words):
    parts = [text]
    for word in words:
        new_parts = []
        for part in parts:
            new_parts.extend(split_by_word(part, word))
        parts = new_parts
    return parts

# 단어 기준으로 문자열 분리하는 기본 함수 정의
def split_by_word(text, word):
    parts = []
    start = 0
    while True:
        index = text.find(word, start)
        if index == -1:
            break
        parts.append(text[start:index + len(word)])
        start = index + len(word)
    parts.append(text[start:])
    return parts

# 여러 단어 기준으로 분리할 단어 리스트 정의
words_to_split = ['빠르고', '스럽고', ',']

# split_text 열을 분리하여 새로운 열 생성
df2 = data.explode(['kss_split_text'])
#df2['all_split_text'] = df2['kss_split_text'].apply(lambda x: split_by_multiple_words(x, words_to_split))
df2['kss_split_text'] = df2['kss_split_text'].apply(lambda x: split_by_multiple_words(x, words_to_split))
df2 = df2.explode(['kss_split_text'])
df2.head(3)
```

```python
#### 위 함수 강화 버전 (고유 단어 개수 기반 추가 필터링)

def text_again_dev(text):
  text = str(text)
  words = text.split()

  # 중복 단어 케이스 (True: 중복 없음)
  result1 = (len(words)==len(set(words)))

  # 연속 단어 케이스
  result2 = 0
  for i in range(1, len(words)):
    if words[i] == words[i-1]:
      result2 = 1
    else:
      result2 = 0

  # 두 케이스가 모두 만족하는 경우 1 반환, 아니면 0 반환 (중복 & 연속)
  result3 = 0
  if (str(result1)=='False') & (result2==1):
    result3 = 1
  else:
    result3 = 0

  #### 세번째 케이스 (set 개수)
  set_len = len(set(words))
  if (result3 == 1) & (set_len <= 3):
    return 1
  else :
    return 0
```

```python
#### 위 함수 강화 버전 (고유 단어 개수 기반 추가 필터링)

def text_again_dev(text):
  text = str(text)
  words = text.split()

  # 중복 단어 케이스 (True: 중복 없음) -> 고유단어 기준
  result1 = (len(words)==len(set(words)))

  # 연속 단어 케이스 -> 이전 단어와 다음 단어가 같은지 확인
  result2 = 0
  for i in range(1, len(words)):
    if words[i] == words[i-1]:
      result2 = 1
    else:
      result2 = 0

  # 두 케이스가 모두 만족하는 경우 1 반환, 아니면 0 반환 (중복 & 연속)
  result3 = 0
  if (str(result1)=='False') & (result2==1):
    result3 = 1
  else:
    result3 = 0

  #### 세번째 케이스 (set 개수)
  set_len = len(set(words))
  if (result3 == 1) & (set_len <= 3):
    return 1
  else :
    return 0

# 해당 코드는 띄워쓰기가 되어있지 않고 연결되면 적용이 안되는 문제점이 있긴 함
# 따라서 replace로 그러한 케이스를 수정해주거나 뒷단에서 수정해줘야 함
```

```python
import kss
from kss import split_sentences

def kss_sentence_split(text):
    return kss.split_sentences(text)
```

```python
#### 단순 일반 구분해서 token화 대상 지정 
data['tokenized_sentence'] = ''
data.loc[data['review_type']=='general','tokenized_sentence'] = data.loc[data['review_type']=='general']['content_lst'].apply(kss_sentence_split)
data.loc[data['review_type']=='simple','tokenized_sentence'] = data.loc[data['review_type']=='simple']['content_lst']

#### 단순은 text, 일반은 리스트 형태로 통일 (리스트로 반환)
data['tokenized_sentence'] = data['tokenized_sentence'].apply(lambda x : x if isinstance(x,list) else [x])
data.head(3)
```

```python
# 여러 단어 기준으로 문자열 분리하는 함수 정의
def split_by_multiple_words(text, words):
    parts = [text]
    for word in words:
        new_parts = []
        for part in parts:
            new_parts.extend(split_by_word(part, word))
        parts = new_parts
    return parts

# 단어 기준으로 문자열 분리하는 기본 함수 정의
def split_by_word(text, word):
    parts = []
    start = 0
    while True:
        index = text.find(word, start)
        if index == -1:
            break
        parts.append(text[start:index + len(word)])
        start = index + len(word)
    parts.append(text[start:])
    return parts

# 여러 단어 기준으로 분리할 단어 리스트 정의
words_to_split = ['빠르고', '스럽고', ',']

# split_text 열을 분리하여 새로운 열 생성
data = data.explode(['tokenized_sentence'])
data['tokenized_sentence'] = data['tokenized_sentence'].apply(lambda x: split_by_multiple_words(str(x), words_to_split))
data = data.explode(['tokenized_sentence'])
data.head(3)
```
